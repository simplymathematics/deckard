```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
df = pd.read_csv("data/merged_afr.csv")


# Melt the ben_survival_time and adv_survival_time, the prediction time, and the attack time, into two columns
# One column for the time and one column for the type of time
interesting_cols = [
    "ben_survival_time",
    "adv_survival_time",
    "predict_time",
    "train_time",
    "adv_fit_time"
]
id_vars = df.columns.difference(interesting_cols)
time_melt = pd.melt(df, id_vars=id_vars, value_vars=interesting_cols,
             var_name="time_type", value_name="time")
# Create a histogram of the time
time_graph = sns.displot(time_melt, x="time", hue="time_type", log_scale=True, kind='kde', )

```

```python
# Look at accuracy, adv_accuracy, log_loss, and adv_log_loss
interesting_cols = [
    "accuracy",
    "adv_accuracy",
    # "log_loss",
    # "adv_log_loss"
]
id_vars = df.columns.difference(interesting_cols)
metric_melt = pd.melt(df, id_vars=id_vars, value_vars=interesting_cols,
             var_name="metric_type", value_name="metric")
# Create a histogram of the metrics
metric_graph = sns.displot(metric_melt, x="metric", hue="metric_type",  kind='kde')
# metric_graph.set(xscale="log")
```
